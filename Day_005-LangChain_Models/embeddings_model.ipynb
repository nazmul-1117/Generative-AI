# Import Libraries

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

# Embeddings Model

```python
embedding = HuggingFaceEmbeddings(
    model_name = 'sentence-transformers/all-MiniLM-L6-v2'
)
```

# Make a Sentence Embeddings

```python
text = "Plantive is my First Big Project"
embed_single_query = embedding.embed_query(text=text)
embed_single_query
```

```python
len(embed_single_query)
```

# Make a Documents embeddings

```python
doc = [
    'Hello World',
    'I love Old Gopal Bhaar',
    'Dhur TaTa'
]
embed_doc_query = embedding.embed_documents(texts=doc)
embed_doc_query
```

```python
print(len(embed_doc_query))
print(len(embed_doc_query[0]))
```
