# AI Model Build

```python
class MyLLMModel:
  def __init__(self):
    print('Custom AI Chat Model Created')

  def predict(self, prompts: str):

    answer_list = [
        'LangChain provides the engineering platform and open source frameworks',
        'Bangladesh is a unitary parliamentary republic based on the Westminster system',
        'Gopal Bhar is appointed as a jester in Raja Krishnachandras court in medieval Bengal',
        "Ixora is a genus of vibrant, tropical evergreen shrubs"
    ]

    return {
        'query': prompts,
        'content': random.choice(answer_list)
    }
```

```python
a = MyLLMModel()
```

```python
a.predict('how are ypu')
```

# Custom Prompt Template

```python
class MyPromptTemplate:
  def __init__(self, template: str, input_variables: list[str]) -> None:
    self.template = template
    self.input_variables = input_variables

  def format(self, input_dict: dict):
    return self.template.format(**input_dict)
```

```python
prompt_template = MyPromptTemplate(
    template= "Hello {length} World {topic}",
    input_variables = ['length', 'topic']
)
```

```python
prompt = prompt_template.format({'length': 10, 'topic': 'AI'})
prompt
```

```python
a.predict(prompt)
```

# My LLM Chain

```python
class MyLLMChain:
  def __init__(self, llm, template) -> None:
    self.llm = llm
    self.template = template

  def run(self, input_dict):
    prompt = self.template.format(input_dict)
    result = self.llm.predict(prompt)

    return result
```

# Build Chains

```python
model = MyLLMModel()
```

```python
template = MyPromptTemplate(
    template= "Hello {length} World {topic}",
    input_variables = ['length', 'topic']
)
```

```python
# Build Chain

chain = MyLLMChain(
    llm = model,
    template = template
  )
```

```python
chain.run({'length': 10, 'topic': 'AI'})
```
